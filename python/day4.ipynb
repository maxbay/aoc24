```python
import numpy as np

with open("../data/day4_dat.txt", 'r') as file:
    lines = [line.strip() for line in file.readlines()]  

grid = np.array([list(line) for line in lines])


# Part 1
NROWS = len(lines)
NCOLS = len(lines[0])

neighbor_kernel = np.array([[i, j] for i in [-1, 0, 1] for j in [-1, 0, 1]])

def return_valid_neighbors(origin, kernel, letter, direction=None):
    neighbors = origin + (np.array([direction]) if direction is not None else kernel)
    
    # mask to filter invalid neighbors
    valid_mask = ~(neighbors < 0).any(axis=1) & ~(neighbors >= NROWS).any(axis=1) & ~(neighbors >= NCOLS).any(axis=1)
    
    # filter on valid idx and letter template
    return np.array([[i, j] for i, j in neighbors[valid_mask] if lines[i][j] == letter])

def find_pattern(origin, pattern, direction=None):
    if not pattern:
        return 1
    
    kernel = neighbor_kernel if direction is None else None
    valid_neighbors = return_valid_neighbors(origin, kernel, pattern[0], direction)
    
    # constrain search to initial direction
    return sum(find_pattern(neighbor, pattern[1:], neighbor - origin if direction is None else direction) for neighbor in valid_neighbors)

n_obs = sum(find_pattern(np.array([row_n, col_n]), 'MAS') for row_n in range(NROWS) for col_n in range(NCOLS) if lines[row_n][col_n] == 'X')

print("Part 1:")
print(n_obs)


# Part 2
def determine_valid(origin = None, kernel = None):
    neighbors = origin + kernel 
    
    mask1 = ~(neighbors < 0).any(axis=1)
    mask2 = ~(neighbors >= NROWS).any(axis=1)
    mask3 = ~(neighbors >= NCOLS).any(axis=1)
    final_mask = mask1 & mask2 & mask3
    
    valid_neighbors = neighbors[final_mask]


    return len(valid_neighbors) == 9

valid_arrangements = np.array(
    [
        [
            ['M','.','M'],
            ['.','A','.'],
            ['S','.','S']
        ],
        [
            ['M','.','S'],
            ['.','A','.'],
            ['M','.','S']
        ],
        [
            ['S','.','M'],
            ['.','A','.'],
            ['S','.','M']
        ],
        [
            ['S','.','S'],
            ['.','A','.'],
            ['M','.','M']
        ]
    ]
)

n_obs = 0
for row_n in range(1, NROWS-1):
    for col_n in range(1, NCOLS-1):
        candidate = lines[row_n][col_n]
        origin = np.array([row_n,col_n])
        
        if candidate == 'A':

            #if determine_valid(origin, neighbor_kernel):
            neighbors = grid[(origin[0]-1):(origin[0]+2),(origin[1]-1):(origin[1]+2)].copy()

            neighbors[1,0] = '.'
            neighbors[0,1] = '.'
            neighbors[2,1] = '.'
            neighbors[1,2] = '.'

            for i in range(4):
                if (neighbors == valid_arrangements[i]).all():
                    n_obs += 1
                    break
            
print("Part 2:")
print(n_obs)


            
        
```
